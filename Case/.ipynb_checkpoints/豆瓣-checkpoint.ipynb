以下几个头部信息通常是必不可少的：

User-Agent：用于标识发送请求的客户端应用程序的信息，包括操作系统、浏览器版本等。有些网站可能会根据User-Agent来识别爬虫，因此需要模拟真实的浏览器。

Accept：指定客户端能够接收的内容类型，一般设置为application/json或text/html等。

Accept-Encoding：指定客户端能够接受的内容编码方式，例如gzip、deflate等。

Accept-Language：指定客户端的首选语言，服务器可以据此返回相应语言的内容。

Referer：指示请求的来源页面，有些网站可能会根据Referer来进行防盗链处理。

Connection：指定连接方式，一般设置为keep-alive表示保持连接。

Host：指定请求的主机地址。

Origin：指定请求的来源地址。

```python
from lxml import etree
url = 'https://movie.douban.com/top250'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36',
}
res1 = requests.get(url, headers=headers, timeout=5)
print(response.headers['Content-Type'])
# 需要的数据在标签中，用lxml
html=etree.HTML(res1.text)
html.xpath('''/html[@class='ua-windows ua-webkit']/body/div[@id='wrapper']/div[@id='content']/div[@class='grid-16-8 clearfix']/div[@class='article']/ol[@class='grid_view']/li[1]/div[@class='item']/div[@class='info']/div[@class='hd']/a/span[@class='title'][1]''')
# 找位置，有点麻烦
name = html.xpath('//div[@class="hd"]/a/span[1]/text()')
score = html.xpath("//div[@id='content']/div[@class='grid-16-8 clearfix']/div[@class='article']/ol[@class='grid_view']/li[*]/div[@class='item']/div[@class='info']/div[@class='bd']/div[@class='star']/span[@class='rating_num']/text()")
# 装包
list(zip(name,score))
```

# 豆瓣250翻页

```python
url = 'https://movie.douban.com/top250'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36',
}
name = []
score = []
for i in range(10):
    para = {
        'start':25*i
    }
    res = requests.get(url, headers=headers, params=para, timeout=5)
    # 需要的数据在标签中，用lxml
    html=etree.HTML(res.text)
    name.extend(html.xpath('//div[@class="hd"]/a/span[1]/text()'))
    score.extend(html.xpath("//div[@id='content']/div[@class='grid-16-8 clearfix']/div[@class='article']/ol[@class='grid_view']/li[*]/div[@class='item']/div[@class='info']/div[@class='bd']/div[@class='star']/span[@class='rating_num']/text()"))
# 装包
list(zip(name,score))
```
